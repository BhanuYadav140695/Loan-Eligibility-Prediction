```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('loan-train.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
Features_with_NaN = [feature for feature in df.columns if df[feature].isnull().sum()!=0]

print(Features_with_NaN)
```

```python
for feature in Features_with_NaN:
    print(feature,'',df[feature].dtypes)
```

```python
for feature in Features_with_NaN:
    if df[feature].dtypes == 'O':
        df[feature] = np.where(df[feature].isnull() , df[feature].mode(),df[feature])
    else:
        df[feature] = np.where(np.nan , df[feature].mean(),df[feature])
```

```python
df.head(30)
```

```python
df.isnull().sum()
```

```python
# Finding How many Categorical Data is Present in data Set

Categorical_Data =[feature for feature in df.columns if df[feature].dtypes == 'O']

print(Categorical_Data)
```

```python
for feature in Features_with_NaN:
    if df[feature].dtypes != 'O':
        pass
    else:
        print(feature, '' , df[feature].unique())
```

```python
df.replace({'Male':1,'Female':0},inplace=True) # Here we use the simple one hot encoding 
```

```python
df.replace({'No':1,'Yes':2},inplace=True) # Here we use the 1 for Yes and 2 for No
```

```python

Dictionary = {'0':1,'1':2,'2':3,'3+':4}

df['Dependents'] = df['Dependents'].map(Dictionary)
```

```python
df.head(20)
```

```python
df['Education'] = np.where(df['Education'] == 'Graduate' , 1,0)
```

```python
df.head()
```

```python
df['Property_Area'].unique()
```

```python
Dictionary = df['Property_Area'].value_counts().to_dict()
print(Dictionary)
```

```python
df['Property_Area'] = df['Property_Area'].map(Dictionary)
```

```python
df=df.drop(['Loan_ID'],axis=1)
```

```python
df.head(50)
```

```python
df.shape
```

```python
X = df.drop(['Loan_Status'],axis=1).values
Y = df['Loan_Status'].values
```

```python
import lazypredict
from lazypredict.Supervised import LazyClassifier

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=10)
```

# Model Prediction

Using Third party library we find the best model fit for this project

```python
clp = LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)
model,prediction = clp.fit(X_train,X_test,Y_train,Y_test)
```

# Model Building 

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold

kfold = KFold(n_splits=3,random_state=8) # This is use for cross validation 
                                         # Which is called Kfold Validation

base_model = DecisionTreeClassifier()    # Decision Tree is use for classification

num_tree = 500

model = BaggingClassifier(base_estimator=base_model,n_estimators=num_tree,random_state=8)
model.fit(X_train,Y_train)
```

```python
Y_hat = model.predict(X_test)
```

```python
model.predict([[1,2,3,1,2,5849,0.0,146.412162,342.0,0.842199,202]])
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_hat)*100)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test,Y_hat)

print(cm)
```
